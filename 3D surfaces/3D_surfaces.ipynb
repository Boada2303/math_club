```python
import sympy as sp
from sympy.abc import x
from sympy import sin, fourier_series, pi, log, power, ln, exp, root
from sympy.utilities.lambdify import lambdify
import numpy as np
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.mplot3d import Axes3D

def plot3d(Z):
    x = np.linspace(-5, 5, 50)
    y = np.linspace(-5, 5, 50)
    X, Y = np.meshgrid(x, y)


    figure = plt.figure(1, figsize = (12, 4))
    subplot3d = plt.subplot(111, projection='3d')

    surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.nipy_spectral, linewidth=0.1)
    plt.show()


```

```python
plot3d(X+Y)
plot3d(X)
plot3d(Y)
```

```python
plot3d(X**2)
plot3d(Y**2)
```

```python
plot3d(np.sin(X+Y))
plot3d(np.sin(X*Y))
plot3d(np.sin(np.sqrt(X**2+Y**2)))
```

```python
plot3d(Z)
```

```python
plot3d(np.sin(X**2))
plot3d(np.sin(Y**2))
plot3d(np.sin(Z**2))
plot3d(Z)
```

```python
plot3d(np.exp(-(X**2+Y**2)))
plot3d(X*np.exp(-(X**2+Y**2)))
plot3d(Y*np.exp(-(X**2+Y**2)))
plot3d(X*Y*np.exp(-(X**2+Y**2)))
```
