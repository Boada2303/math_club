```python
import sympy
import numpy as np
import matplotlib.pyplot as plt
import math

guess =3 #initial guess, can be any value.
cycles = 7
a = 1        #counter

print ("initial guess: "+str(guess))
guesses = [guess]

while(a < cycles): 
    
    function = 2**guess-guess**2                #manually adding the function using "guess" as the variable for the equation                  
    derivative = 2**guess*math.log(2) - 2*guess #manually adding the derivative 
    
    guess = (guess - (function/derivative))     # formula to get the next guess
    print (str(a) + ": " + str(guess))
    guesses.append(guess)   
    a = a+1
```

```python
#plot guesses
plt.plot(guesses)
plt.grid(True)
plt.show()
```
