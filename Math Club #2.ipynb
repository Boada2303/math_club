```python
import sympy as sp
from sympy.abc import x
from sympy import sin, fourier_series, pi, log, power, ln, exp, root
from sympy.utilities.lambdify import lambdify
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fourierSeries(func, n):
    s = fourier_series(func, (x, -4, 10))
    FS = s.truncate(n)

    #print(FS)
    t = np.linspace(-4, 10, 1000)

    evalfunc = lambdify(x, func, modules=['numpy'])
    evaltaylor = lambdify(x, FS, modules=['numpy'])
    
    
    plt.figure(figsize=(20,10))
    #plt.axhline(1.16, color="gray")

    plt.grid(True)
    plt.plot(t, evalfunc(t), 'b', label='function')
    plt.plot(t, evaltaylor(t), 'r', label='Fourier Series')
    plt.legend(loc='best')
  
    plt.show()
```

```python
fourierSeries(x,4)
fourierSeries(x,9)
```

```python
fourierSeries(x**2, 7)
```

```python
fourierSeries(x**3, 8)
```

```python
fourierSeries(exp(x), 8)
```

```python
func = sp.Piecewise(
    (0, x < -3),
    (0, x > 3),
    (1, True ))

fourierSeries(func, 7)
fourierSeries(func, 20)
fourierSeries(func, 60)
```
